```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.impute import SimpleImputer
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE

# Load the data into a DataFrame
# df = pd.read_csv('your_data.csv')

# Select the features and target variable
target = 'IS_HCP'
features = df.columns.drop(target)
X = df[features]
y = df[target]

# Drop rows with missing values in the target variable
df = df.dropna()

# Check for NaN values in the target variable after dropping rows
if y.isnull().sum() > 0:
    print("NaN values present in the target variable. Please handle them before proceeding.")
else:
    # Impute missing values in the features
    imputer = SimpleImputer(strategy='mean')
    X_imputed = imputer.fit_transform(X)

    # Select the top k features using chi-squared test
    k = len(features)
    selector = SelectKBest(score_func=chi2, k=10)
    X_selected = selector.fit_transform(X_imputed, y)
    selected_feature_indices = selector.get_support(indices=True)
    selected_features = features[selected_feature_indices]
    y_imputed = imputer.fit_transform(y.values.reshape(-1, 1))
    y = pd.Series(y_imputed.flatten(), name=target)
    # Create a new DataFrame with the selected features and target variable
    df_selected = pd.DataFrame(X_selected, columns=selected_features)
    df_selected[target] =y
    mean_value=df_selected[target].mean()
    df_selected[target].fillna(mean_value)
    df_selected[target].dropna()
    # Check for NaN values in the target variable before fitting the model
    if df_selected[target].isnull().sum() > 0:
        print("NaN values still present in the target variable. Please handle them before proceeding.")
    else:

        from xgboost import XGBRFClassifier
        from sklearn.ensemble import RandomForestClassifier
        from keras.models import Sequential
        from keras.layers import Dense
        from sklearn.preprocessing import StandardScaler

        from sklearn.preprocessing import StandardScaler
        scaler=StandardScaler()
        X_train_scaled =scaler.fit_transform(df[selected_features])
        X_test_scaled =scaler.fit_transform(data[selected_features])
        oversampler = SMOTE()
        X_train_balanced, y_train_balanced = oversampler.fit_resample(X_train_scaled, df[target])
        model = RandomForestClassifier()
        param_grid = {
                  'n_estimators': [100, 200, 300],
                  'max_depth': [5, 10, 15],
                  'min_samples_split': [2, 5, 10]
                    }
        grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=5)
        grid_search.fit(X_train_balanced, y_train_balanced)
        best_params = grid_search.best_params_

        rf_classifier_best = RandomForestClassifier(**best_params)
        rf_classifier_best.fit(X_train_balanced, y_train_balanced)
        prediction = rf_classifier_best.predict(X_test_scaled)


        df_test_predicted = pd.DataFrame({'Predicted_IS_HCP': prediction})
        df_test_with_predictions = pd.concat([data, df_test_predicted], axis=1)
```
